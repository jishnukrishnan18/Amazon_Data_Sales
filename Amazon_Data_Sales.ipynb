DATA ANALYSIS AND VISUALIZATION

```python
#1.Which month has the highest sales?
#2.Which product category generates the highest revenue?
#3.Are discounted products generating more total revenue than non-discounted products?
#4.Which region generates the highest total revenue?
#5.Which product category has the highest number of reviews?
#6.How are product ratings distributed across all products?
#7.What percentage of total revenue comes from each product category?
#8.Which customer region purchases the highest quantity of products?
#9.What variables in the dataset are strongly related to each other?
#10.Which product category contributes the least revenue?
#11.Which payment method dominates in each product category?
#12 Which cities or states have the highest number of customers?
#13.How many orders occur each month?
#14.How does product price vary across categories?
#15.How frequently do regions place orders?
```

```python
#1.Which month has the highest sales?
df['Month']=df['Order_Date'].dt.month_name().sort_values(ascending=False)
sales=df.groupby('Month')['Total_Revenue'].sum()
sales.head(1)

sales.plot(kind='bar')

plt.title('Total Revenue by Month')
plt.xlabel('Month')
plt.ylabel('Total Revenue')

plt.show()
```

```python
#2.Which product category generates the highest revenue?

df.groupby('Product_Category')['Total_Revenue'].sum().sort_values(ascending=False)
```

```python
#3.Are discounted products generating more total revenue than non-discounted products?

df['Discounted'] = df['Discount_Percent'] > 0
discounted_revenue = df.groupby('Discounted')['Total_Revenue'].sum()

print(discounted_revenue.sort_values(ascending=False))

sns.barplot(x=discounted_revenue.index, y=discounted_revenue.values, color='red')

plt.title('Revenue Comparison: Discounted vs Non-Discounted Products')
plt.xlabel('Discount Applied')
plt.ylabel('Total Revenue')
plt.show()
```

```python
#4.Which region generates the highest total revenue?

region=df.groupby('Customer_Region')['Total_Revenue'].sum().sort_values(ascending=False)
print(region)
region.plot(kind='barh')
plt.title('Total Revenue by Customer Region')
plt.xlabel('Customer Region')
plt.ylabel('Total Revenue')
plt.show()
```

```python
#5.Which product category has the highest number of reviews?

df.groupby('Product_Category')['Review_Count'].sum().sort_values(ascending=False)
```

```python
#6How are product ratings distributed across all products?
df['Rating'].plot(kind='hist',edgecolor="white")
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
#7.What percentage of total revenue comes from each product category?

df.groupby('Product_Category')['Total_Revenue'].sum().plot(kind='pie',autopct='%1.1f%%')

plt.title('Revenue Share by Product Category')
plt.ylabel('')
plt.show()
```

```python
#8.Which customer region purchases the highest quantity of products?

df.groupby('Customer_Region')['Quantity_Sold'].sum().plot(kind='bar')

plt.title("Quantity Sold by Region")
plt.xlabel("Region")
plt.ylabel("Quantity Sold")
plt.show()
```

```python
#9.What variables in the dataset are strongly related to each other?

corr = df[['Price','Discount_Percent','Quantity_Sold','Rating','Review_Count','Discounted_Price','Total_Revenue']].corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#10.Which product category contributes the least revenue?

category_sales = df.groupby('Product_Category')['Total_Revenue'].sum().sort_values()
print(category_sales.head(1))  
category_sales.plot(kind='bar')
plt.title('Revenue by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.show()
```

```python
#11.Which payment method dominates in each product category?

data = df.groupby(['Product_Category', 'Payment_Method']).size().unstack()
data.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title("Payment Method Distribution by Category")
plt.xlabel("product Category")
plt.ylabel("Number of Orders")
plt.legend(title="Payment Method")
plt.show()
```

```python
#12.Which cities or states have the highest number of customers?

city_counts = df['Customer_Region'].value_counts().head(10)
city_counts.plot(kind='barh', figsize=(10,5))
print(city_counts)
plt.title("Top 10 Cities by Number of Customers")
plt.xlabel("City")
plt.ylabel("Number of Customers")
plt.xticks(rotation=45)
plt.show()
```

```python
#13.How many orders occur each month?

df['Month'] = pd.to_datetime(df['Order_Date']).dt.month
df.groupby('Month')['Order_Id'].count().sort_index().plot()
plt.xlabel("Month")
plt.ylabel("Orders")
plt.title("Monthly Orders (Ordered)")
plt.show()
```

```python
#14 How does product price vary across categories?

sns.boxplot(x='Product_Category', y='Price', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
#15.How frequently do regions place orders?

region_counts = df['Customer_Region'].value_counts()
print(region_counts)
region_counts.plot(kind='bar')
plt.title("Order Frequency by Region")
plt.xlabel("Region")
plt.ylabel("Number of Orders")

plt.show()
```

CONCLUSION


Exploratory Data Analysis (EDA) was performed on the Amazon sales dataset to understand sales patterns, customer behavior, and product performance. The dataset included features such as product category, price, discount, quantity sold, customer region, ratings, and revenue.

The analysis revealed that sales vary over time, indicating seasonal trends and peak periods. Certain product categories contributed more to total revenue, showing their importance in business growth. Customer behavior also differed by region, with some areas generating higher orders and revenue.

A relationship between price and quantity sold suggested that demand is somewhat price-sensitive. Discount analysis showed that while discounts can increase sales, they do not always lead to higher revenue, highlighting the need for optimized pricing strategies. Additionally, products with higher ratings and more reviews performed better, emphasizing the importance of customer feedback.

Overall, EDA helped uncover key insights that can support data-driven decision-making, improve customer satisfaction, and enhance overall business performance.
