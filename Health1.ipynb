LOGISTIC REGRESSION


```python
df.columns
```

```python
feature_col=['Age', 'Gender', 'Daily_Social_Media_Hours', 'Platform_Usage',
       'Sleep_Hours', 'Screen_Time_Before_Sleep', 'Academic_Performance',
       'Physical_Activity', 'Social_Interaction_Level', 'Stress_Level',
       'Anxiety_Level']
logreg_x=df[feature_col]
logreg_y=df['Depression_Risk']
```

```python
logreg_x=pd.get_dummies(logreg_x,drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

logreg_x_train, logreg_x_test, logreg_y_train, logreg_y_test = train_test_split(
    logreg_x, logreg_y, test_size=0.2, random_state=0
)

```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
logreg_x_train = scaler.fit_transform(logreg_x_train)
logreg_x_test = scaler.transform(logreg_x_test)
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression(solver='lbfgs',max_iter=1000)
logreg.fit(logreg_x_train,logreg_y_train)
logreg_pred=logreg.predict(logreg_x_test)
```

```python
from sklearn import metrics
print("accuracy",metrics.accuracy_score(logreg_y_test,logreg_pred)*100)
```

```python
# Hypertuning of parameters for the best fit model
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression())
])

param_grid = {
    'model__C': [0.1, 1, 10],
    'model__penalty': ['l2'],
    'model__solver': ['lbfgs','saga'],
    'model__max_iter': [1000, 3000, 5000]
}

grid = GridSearchCV(pipe, param_grid, cv=5, n_jobs=-1)
grid.fit(logreg_x_train, logreg_y_train)

print("Best params:", grid.best_params_)

y_pred = grid.predict(logreg_x_test)

from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(logreg_y_test, y_pred))
```

After Hypertuning

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression(solver='saga',max_iter=1000,C=10,penalty='l2')
logreg.fit(logreg_x_train,logreg_y_train)
log_pred=logreg.predict(logreg_x_test)
from sklearn import metrics
accuracy= metrics.accuracy_score(logreg_y_test, log_pred)*100
print("Accuracy =",accuracy,"%")
print("\nConfusion Matrix:")
print(metrics.confusion_matrix(logreg_y_test, log_pred))

print("\nClassification Report:")
print(metrics.classification_report(logreg_y_test, log_pred))
```

## CONCLUSION
- After performing Exploratory Data Analysis (EDA), Classification Modeling, and Hyperparameter Tuning, the key findings are summarized below:

### 1. Data Understanding & Preprocessing
- The dataset contained both numerical and categorical features.
- Categorical variables were converted into numerical form using get dummies function.
- Dataset was split into training and testing sets for proper evaluation

### 2. Exploratory Data Analysis (EDA)
- Teenagers spend a significant amount of time on social media daily.
- Certain platforms are more popular and widely used among teens.
- Sleep duration varies, and higher screen time before bed may reduce sleep quality.
- Increased social media usage is linked with higher stress and anxiety levels.
- Academic pressure also contributes to stress among teenagers.
- Social media usage and stress levels differ across gender.
- EDA helped identify behavioral trends, mental health patterns, and screen-time habits among teens.

### 3. Classification Models Applied
- The following models were implemented and evaluated:
  - Logistic Regression
  - K-Nearest Neighbors (KNN)
  - Support Vector Machine (SVM)
  - Naive Bayes
  - Decision Tree
  - Random Forest

### 4. Model Evaluation
- Metrics used:
  - Accuracy
  - Precision, Recall, F1-score
### 5. Hyperparameter Tuning
- Applied GridSearchCV to optimize models.
- Improved model stability and generalization.
- Observed that:
  - Large parameter grids increase computation time significantly
  - Efficient tuning strategies are important for performance vs speed trade-off
### 6. Best Model
- Based on test results:
  - Decision Tree Classifier achieved the highest accuracy among all models
### 7. Final Conclusion
- EDA revealed that social media usage, sleep habits, academic pressure, and screen time influence teen stress and anxiety levels.
- Multiple classification models such as Logistic Regression, KNN, SVM, Naive Bayes, Decision Tree, and Random Forest were trained and evaluated using accuracy and classification metrics.
- Hyperparameter tuning improved model performance, and the Decision Tree model achieved the best overall accuracy for prediction.
